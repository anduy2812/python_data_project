# Plotting Using Pandas

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

### Plotting Series

```python
job_counts = df['job_title_short'].value_counts()
job_counts
```

```python
# Plotting
job_counts.plot(kind='bar')

# Labeling
plt.xlabel('')
plt.ylabel('Count of Postings')
plt.title('Job Postings by Job Title')
plt.xticks(rotation=45, ha='right') 
plt.show()

plt.show()
```

### Plotting DataFrames

```python
df.plot(kind='line', y='salary_year_avg', x='job_posted_date')
plt.show()
```

**The problem with DataFrames (at least large ones) is data still may need to be cleaned up**

```python
# Sorting DataFrames by job posted date
df = df.set_index('job_posted_date')
df.sort_index(inplace=True)
df.head()
```

```python
df.plot(y='salary_year_avg', kind='line')
plt.show()
```
