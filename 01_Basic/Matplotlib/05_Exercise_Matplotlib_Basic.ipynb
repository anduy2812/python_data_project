### Question
Graph the average salary (`salary_year_avg`) by `job_title_short` as a horizontal bar chart. Order it from the highest salary to the lowest. Include a title, and labels for the x & y-axis.

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
job_salary = df.groupby('job_title_short')['salary_year_avg'].median()

job_salary = job_salary.sort_values()

job_salary
```

```python
# Plotting
job_salary.plot(kind='barh')

# Labeling
plt.xlabel('Yearly Salary ($USD)')
plt.ylabel('')
plt.title('Average Yearly Salary by Job Title')
```
